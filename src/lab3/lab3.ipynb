```python
import numpy as np
training_data = [
  ("The dog ate the apple".split(), ["determiner", "noun", "verb", "determiner", "noun"]),
  ("Everybody read that book".split(), ["noun", "verb", "determiner", "noun"]),
  ("Everybody does machine learning nowadays".split(), ["noun", "verb", "noun", "noun", "adverb"]),
]

tag_to_ind = {
  "determiner": 0,
  "noun": 1,
  "verb": 2,
  "adverb": 3,
}

ind_to_tag = {v: k for k, v in tag_to_ind.items()}
vocab = {a: b for b, a in enumerate(set(word for sentence, _ in training_data for word in sentence))}
vocab
```

```python
import lightning as L
import torch

class LSTMTagger(L.LightningModule):
  def __init__(self, vocab, tag_to_ind):
    super().__init__()
    self.model =  torch.nn.Sequential(
      torch.nn.Embedding(len(vocab), embedding_dim=6), # word index -> embedding vector (6x1)
      torch.nn.LSTM(input_size=6, hidden_size=12),
      torch.nn.Linear(in_features=12, out_features=len(tag_to_ind)),
      torch.nn.LogSoftmax(dim=1) 
    )
  def training_step(self, batch, batch_idx):
    x, _ = batch
    y = self.model(x)
    loss = torch.exp(torch.nn.functional.cross_entropy(y, _))
    return loss

  def test_step(self, batch, batch_idx):
    x, _ = batch
    y = self.model(x)
    loss = torch.exp(torch.nn.functional.cross_entropy(y, _))
    return loss

lstm = LSTMTagger(vocab, tag_to_ind)
lstm
```
